```python
#Homework 6 yipppeeeee
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def sin(x,A,w):
    return A*np.sin(w*x)
```

```python
x_data = np.linspace(0,2*np.pi,1000)
A = np.array([1,2,3,4,5])
w = np.array([1,2,3,4,5])

for i in range(len(A)):
    print(A[i])
```

```python
plt.figure()
plt.title('A Chaotic graph of many different Sinusoids :)')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
for i in range(len(A)):
    y = sin(x_data,A[i],w[i])
    plt.scatter(x_data,y)
plt.show()
```

```python
#Step one generate two 40-element lists of random numbers from 0-100
randomList = []
def createRandList(size):
    list1 = []
    for i in range(0,size):
        list1.append(random.random()*100)
    return list1
rand1 = createRandList(40)
rand2 = createRandList(40)
x = list(range(0,40))
```

```python
plt.figure()
plt.plot(figsize = 1000)

plt.plot(x,rand1, color = 'orange',linewidth = 10, label = 'this is random')
plt.plot(x,rand2, color = 'red', linestyle = 'dashed', label = 'very very random')
plt.legend()
plt.title('Random graph')
plt.show()
```
